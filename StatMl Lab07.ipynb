```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the Iris dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'target']
data = pd.read_csv(url, names=column_names)
X = data.iloc[:, :-1]
plt.figure(figsize=(12, 8))
for i, feature in enumerate(X.columns):
    plt.subplot(2, 2, i + 1)
    plt.hist(X[feature], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# @title pca analysis
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
data = pd.read_csv(url, names=column_names)
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

# Step 4: Compute the covariance matrix
cov_matrix = np.cov(X_standardized, rowvar=False)

# Step 5: Compute the eigenvectors and eigenvalues
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Step 6: Sort eigenvalues in descending order
eigenvalue_index = np.argsort(eigenvalues)[::-1]
eigenvalues_sorted = eigenvalues[eigenvalue_index]
eigenvectors_sorted = eigenvectors[:, eigenvalue_index]

# Step 7: Choose the top-k eigenvectors (principal components)
k = 2  # You can choose the number of principal components
top_k_eigenvectors = eigenvectors_sorted[:, :k]

# Step 8: Create a new feature matrix using the selected principal components
reduced_data = np.dot(X_standardized, top_k_eigenvectors)

# Step 9: Visualize the reduced-dimensional data
reduced_df = pd.DataFrame(data=reduced_data, columns=["Principal Component 1", "Principal Component 2"])
reduced_df["class"] = y

plt.figure(figsize=(10, 8))
sns.scatterplot(x="Principal Component 1", y="Principal Component 2", hue="class", data=reduced_df, palette="viridis")
plt.title('PCA Reduced Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title="Class")
plt.show()
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df=pd.read_csv(url,names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width','target'])
feature=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
x=df.loc[:, feature].values
y=df.loc[:,['target']].values
x=StandardScaler().fit_transform(x)
```

```python
pca=PCA(n_components=2)
principal_components=pca.fit_transform(x)
principalDataFrame=pd.DataFrame(data=principal_components,columns=['PC1','PC2'])
```

```python
targetDataframe=df[['target']]
newdataframe=pd.concat([principalDataFrame,targetDataframe],axis=1)
newdataframe
```

```python
plt.scatter(principalDataFrame.PC1,principalDataFrame.PC2)
plt.title('pc1 against pc2')
#plt.xlable('PC1')
```

```python
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(1,1,1)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_title('plot of pc1 and pc2',fontsize=20)
targets=['Iris-setosa','Iris-versicolor','Iris-virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets, colors):
    indices_to_keep = newdataframe['target'] == target
    ax.scatter(newdataframe.loc[indices_to_keep, 'PC1'],
                newdataframe.loc[indices_to_keep, 'PC2'],
                c=color, s=50)
ax.legend(targets)
ax.grid()
# Step 4: Apply explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
print(f"Explained Variance Ratio for PC1: {explained_variance_ratio[0]:.2f}")
print(f"Explained Variance Ratio for PC2: {explained_variance_ratio[1]:.2f}")
```
