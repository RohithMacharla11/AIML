```python
import numpy as np
list1 =[5,6,7,8]
list2= [1,2,3,4]
list_type = type(list1)
arr1 = np.array(list1)
arr2 = np.array(list2)
arr_type = type(arr1)
print("list :", list_type)
print("array : ", arr_type)
```

```python
multiply = arr1 * arr2
division = arr1/arr2
power = arr1 ** arr2
print(f"Result of multiplication: {multiply}")
print(f"Result of division: {division}")
print(f"Result of power off: {power}")
```

```python
print(f"Sine : {np.sin(arr1)}")
print(f"Log : {np.log(arr1)}")
print(f"B-10 Log : {np.log10(arr1)}")
print(f"B-2 log : {np.log2(arr1)}")
print(f"Exponent : {np.exp(arr1)}")
```

```python
people = ["Rohith", "Chinnu", "Macharla", "Tharun", "Dheeraj", "kailash"]
age = [21,19,18,17,20,17]
height = [155, 157, 160, 166, 175, 177]
weight = [60, 65, 59, 62, 64, 70]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight,height)
plt.title("Relationship between Height ang Weight of Patients")
plt.xlabel("Weight in kgs")
plt.ylabel("Height in mts")
plt.show()
```

```python
plt.scatter(age,height)
plt.title("Relationship between Age ang Height of Patients")
plt.xlabel("Age of Patients")
plt.ylabel("Height in mts")
plt.show()
```

```python
plt.scatter(age,weight)
plt.title("Relationship between age ang Weight of Patients")
plt.xlabel("Age of Patients")
plt.ylabel("Weight in Kgs")
plt.show()
```

```python
plt.figure(figsize =(12,4))
plt.title("People's Weight in Kgs", fontsize = 16, fontstyle = 'italic')
plt.bar(x=people, height = weight, width = 0.6, color = {'orange', 'red', 'blue', 'green'},edgecolor='brown', alpha= 0.9)
plt.xlabel("People",fontsize = 16)
plt.xticks(fontsize = 14, rotation = 50)
plt.yticks(fontsize = 14)
plt.ylabel("Weight(in Kgs)", fontsize = 16)
plt.show()
```

```python
plt.figsize = (12,5)
plt.hist(weight, color= 'red', edgecolor= 'black', alpha = 0.9, bins =5)
plt.title("Histogram of Patient weight",fontsize = 14)
plt.xlabel("Weight",fontsize = 16)
plt.xticks(fontsize = 14, rotation = 50)
plt.show()
```
