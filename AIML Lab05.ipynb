```python
f=train.groupby('Pclass')['Age'].mean()
f=pd.DataFrame(f)
f.plot.bar(y='Age')
plt.title("Average age in passengers",fontsize=17)
plt.ylabel("Age (years)",fontsize=7)
plt.xlabel("passenger class",fontsize=17)
```

```python
a=list(f['Age'])
def imputs_age(cols):
  Age=cols[0]
  Pclass=cols[1]
  if pd.isnull(Age):
    if Pclass==1:
      return a[0]
    elif Pclass==2:
      return a[1]
    else:
      return a[2]
  else:
    return Age
```

```python
train['Age']=train[['Age','Pclass']].apply(imputs_age,axis=1)
d=train.describe()
dT=d.T
dT.plot.bar(y='count')
plt.title("bar plot of the count of numeric features",fontsize=17)
```
